```python
import google.generativeai as genai
import os

# 1. Configuração com a chave que funcionou (193E)
genai.configure(api_key="AIzaSyBoakPecnu2YkJWwf7uDXqHW3-9HUS193E")

# 2. Leitura dos ficheiros (O teu contexto)
def ler_md(nome):
    if os.path.exists(nome):
        with open(nome, 'r', encoding='utf-8') as f: return f.read()
    return ""

persona = ler_md('persona.md')
cv = ler_md('cv.md')
faq = ler_md('Recruiter_FAQs.md')
bio = ler_md('Biografia.md')

# Montamos o contexto de texto
prompt_sistema = f"{persona}\n\nCONTEXTO DO SIMÃO:\n{cv}\n{faq}\n{bio}"

# 3. Inicialização do Gémeo Digital (Mantendo o 2.5 Flash que deu sorte)
model = genai.GenerativeModel(
    model_name='models/gemini-2.5-flash', 
    system_instruction=prompt_sistema
)

try:
    print("🧠 A carregar a memória do Simão Digital...")
    # Teste de conhecimento
    response = model.generate_content("Quem és tu e qual é o teu objetivo principal?")
    
    print("-" * 30)
    print(f"🤖 GÉMEO DIGITAL DIZ:\n{response.text}")
    print("-" * 30)
    print("✅ MEMÓRIA CARREGADA! Se a resposta estiver correta, avisa-me para metermos a foto.")

except Exception as e:
    print(f"❌ Erro ao carregar memória: {e}")
```

```python
# --- CICLO DE INTERAÇÃO COM O GÉMEO DIGITAL ---

chat = model.start_chat(history=[])

print("🤖 O teu Gémeo Digital está online! (Escreve 'sair' para encerrar)")
print("-" * 50)

while True:
    pergunta_user = input("Tu: ")
    
    if pergunta_user.lower() in ['sair', 'exit', 'quit']:
        print("🤖 Simão Digital: Até à próxima! Foi um gosto.")
        break
        
    try:
        response = chat.send_message(pergunta_user)
        print(f"\n🤖 Simão Digital: {response.text}\n")
        print("-" * 30)
    except Exception as e:
        print(f"❌ Erro na interação: {e}")
        break
```
